```python
# Importar bibliotecas necessrias
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

# Carregar arquivo de imagem
figura01 = 'placa01.jpg'
img = mpimg.imread(figura01)
imgplot = plt.imshow(img)
plt.show()
```

```python
# Importar bibliotecas necessrias
import easyocr

# Definir Portugus como idioma
leitor = easyocr.Reader(['pt'])

# Ler o arquivo de imagem
resultados = leitor.readtext(figura01, paragraph = False)

# Itera sobre o resultado
texto = []
for resultado in resultados:
    texto.append(resultado[1])

print(' '.join(texto)) # Apresenta o texto extrado da imagem ao usurio
```
