```python
# Importar bibliotecas necess치rias
import numpy as np
import cv2

# Carregar classificadores j치 treinados
face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
eye_cascade = cv2.CascadeClassifier('haarcascade_eye.xml')

# Ler arquivo de imagem
imagem = cv2.imread('faces_image02.jpg')

# Converter imagem para escala de cinza
imagemcinza = cv2.cvtColor(imagem, cv2.COLOR_BGR2GRAY)

# Detectar faces na imagem
faces = face_cascade.detectMultiScale(imagemcinza, scaleFactor = 1.3, minNeighbors = 5, minSize = (30, 30))

# Desenhar um ret칙ngulo nas faces encontradas
for (x,y,w,h) in faces:
    cv2.rectangle(imagem, (x, y), (x + w, y + h), (255, 0, 0), 2)
    roi_gray = imagemcinza[y:y+h, x:x+w]
    roi_color = imagem[y:y+h, x:x+w]
    eyes = eye_cascade.detectMultiScale(roi_gray) # Detectar olhos na imagem
    for (ex,ey,ew,eh) in eyes: # Desenhar um ret칙ngulo nos olhos encontrados
        cv2.rectangle(roi_color, (ex, ey), (ex + ew, ey + eh), (0, 255, 0), 2)
```

```python
# Importar bibliotecas necess치rias
import matplotlib.pyplot as plt

# Apresentar faces detectadas na imagem
plt.imshow(imagem)
plt.show()
```
