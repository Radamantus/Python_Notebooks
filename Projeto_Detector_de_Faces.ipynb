```python
# Importar bibliotecas necessárias
import cv2

# Carregar classificador já treinado
modelo = 'haarcascade_frontalface_alt.xml'
# modelo = 'haarcascade_frontalface_default.xml'
classificador = cv2.CascadeClassifier(modelo)

# Ler arquivo de imagem
imagem = cv2.imread('faces_image02.jpg')

# Converter imagem para escala de cinza
imagemcinza = cv2.cvtColor(imagem, cv2.COLOR_BGR2GRAY)

# Detectar faces na imagem
faces = classificador.detectMultiScale(imagemcinza, scaleFactor = 1.1, minNeighbors = 3, minSize = (30, 30))

# Desenhar um retângulo nas faces encontradas
for (x, y, l, a) in faces:
    cv2.rectangle(imagem, (x, y), (x + l, y + a), (0, 255, 0), 3)
    
# Apresentar ao usuário
print('Detecção de faces finalizada.')
```

```python
# Importar bibliotecas necessárias
import matplotlib.pyplot as plt

# Apresentar faces detectadas na imagem
plt.imshow(imagem)
plt.show()
```
